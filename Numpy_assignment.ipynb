```python
#100. How do you detect outliers in a dataset using IQR?
data = np.array([10, 12, 14, 15, 18, 21, 22, 24, 29, 100])
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
l = Q1 - 1.5 * IQR
u = Q3 + 1.5 * IQR

outliers = data[(data < l) | (data > u)]
print(outliers)
```
